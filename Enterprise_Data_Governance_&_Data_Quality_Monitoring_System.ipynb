```python
import pandas as pd
import numpy as np
```

```python
# Load dataset
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
```

```python
# DATA PROFILING
profile = pd.DataFrame({
    "column": df.columns,
    "missing_values": df.isnull().sum().values,
    "missing_pct": (df.isnull().mean()*100).values,
    "dtype": df.dtypes.values
})

print("DATA PROFILE:\n", profile)
```

```python
# DATA QUALITY RULES

# Rule 1: Missing Age imputation flag
df["age_missing_flag"] = df["Age"].isnull().astype(int)

# Rule 2: Duplicate detection
df["duplicate_flag"] = df.duplicated().astype(int)

# Rule 3: Outlier detection (Fare)
q1, q3 = df["Fare"].quantile([0.25, 0.75])
iqr = q3 - q1
df["fare_outlier"] = ((df["Fare"] < (q1 - 1.5*iqr)) | (df["Fare"] > (q3 + 1.5*iqr))).astype(int)
```

```python
# DATA QUALITY SCORE

df["data_quality_score"] = (
    1 - (df["age_missing_flag"] + df["duplicate_flag"] + df["fare_outlier"]) / 3
)

print(df[["Survived", "Age", "Fare", "data_quality_score"]].head())
```

```python
# Export
df.to_csv("data_quality_output.csv", index=False)
```
